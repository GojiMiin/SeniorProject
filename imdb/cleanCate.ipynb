```python
import pandas as pd
```

```python
allData = pd.read_csv('18_19.csv')
allData
```

```python
singleCate = []
allCate = allData["category"]
allCate
```

```python
print(allCate[0].split(','))
```

```python
for i in allCate:
    row = i.split(',')
    for j in row:
        cleanSpace = j.replace(" ", "")
        if cleanSpace not in singleCate:
            singleCate.append(cleanSpace)
```

```python
singleCate.sort()
singleCate
```
